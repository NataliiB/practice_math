**Задача 1** \
Дані — перегляд (хв): \
[3, 4, 5, 6, 7]

Знайти медіану.

```python
arr = [4, 5, 3, 7, 6]
def count_median(arr):
  n = len(arr)
  arr.sort() # sorted 
  if n%2 == 1:
      median = arr[n//2]
  else:
      median = (arr[n//2-1]+arr[n//2])/2
  return median
count_median(arr)
```

Дані: [3, 3, 4, 100] \
Знайти median та mean. Що про них можна сказати?

```python
arr = [3, 3, 4, 100] 

def count_mean(arr):
   sum = 0
   for item in arr:
      sum += item
      mean = sum/len(arr)
   return mean
print(count_median(arr),count_mean(arr))
# У масиві невелика кількість даних і один великий викид, через що середнє значення спотворене, а медіана (3.5) більш репрезентативно показує центр.
```

**Задача 3**

Дані днів користування сервісом: [1, 2, 2, 2, 3, 90]

Пояснити: чому median краще mean?

```python
arr = [1, 2, 2, 2, 3, 90]
def count_median(arr):
  n = len(arr)
  arr.sort() # sorted 
  if n%2 == 1:
      median = arr[n//2]
  else:
      median = (arr[n//2-1]+arr[n//2])/2
  return median

def count_mean(arr):
   sum = 0
   for item in arr:
      sum += item
      mean = sum/len(arr)
   return mean

print(count_median(arr),count_mean(arr))

# Данні мають викид 90 днів. В той час, як медіана стійка до викидів, середнє – ні.
```

**Задача 1** \
Дані: [10, 12, 12, 15, 20]\
Знайти моду.

```python
arr = [10, 12, 12, 15, 20]
def count_mode(arr):
    d = {}
    for i in arr:
        d[i] = arr.count(i)
    mode = [k for k,v in d.items() if v == max(d.values())]
    return mode

count_mode(arr)
```

Задача 2 — два набори з однаковим середнім, але різним розкидом \
A: [4, 5, 6] \
B: [1, 5, 9] 

Середні однакові.  
Який набір стабільніший?

```python
a_set = [4, 5, 6]
b_set = [1, 5, 9]
def count_dispersion(set):
    avg = sum(set)/len(set)
    squares = [(i-avg)**2 for i in set]
    result = sum(squares)/len(set)
    return result
print(count_dispersion(a_set),count_dispersion(b_set))

# b_set має значно більшу дисперсію,тобто елементи сильніше розкидані відносно середнього.

# a_set більш концентрований навколо середнього, тому стабільніший
```

### Std 
Дані: [4, 5, 6, 9, 10]  

Порахуй variance і std. 


```python
def count_variance(data):
    variance = sum([(i- sum(data) / len(data))**2 for i in data])/len(data)
    std_value = variance ** 0.5
    return variance, std_value
count_variance([4, 5, 6, 9, 10])
```
